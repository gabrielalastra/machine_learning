<h3 align='center'>Examples of Data Distributions</h3>

```python
import numpy as np
import matplotlib.pyplot as plt
```

<body>
<h5>Uniform Distribution</h5>
Has equal probability across any given value.</br> we expect to see basically zero outsiders of the range. </br>but between x1 and x2, we expect to see a flat line as a constant

```python
value = np.random.uniform(-10,10,100000)
plt.hist(value,50)
plt.show()
```

<h5>Normal / Gaussian</h5>

```python
from scipy.stats import norm

x = np.arange(-3,3,0.001)
plt.plot(x, norm.pdf(x)) #probability density function for a norm distribution on those x values
```

```python
middle = 8
sigma=4.0 #standard deviation of the data
values = np.random.normal(middle, sigma, 10000) #num of points
plt.hist(values, 50)
plt.show()
```

<body><h5>Exponential PDF / "Power Law"</h5>

</body>

```python
from scipy.stats import expon

x = np.arange(0,10,0.001)
plt.plot(x, expon.pdf(x))
```

<body><h5>Binomial Probability Mass Funciont</h5>
Discrete data
</body>

```python
from scipy.stats import binom
n = 10
p = 0.5
x = np.arange(0,10,0.001)
plt.plot(x, binom.pmf(x, n, p))
```

<body><h5>Poisson Rpobability Mass Fucntion</h5>
if you have any info about the average of number that can happen in some given time period.</br> This could give you a way to predict the odds of getting some other value instead on a given future day
</body>

```python
from scipy.stats import poisson

middle = 500
x = np.arange(400,600,0.5)
plt.plot(x, poisson.pmf(x, middle))
```

<h3 align='center'>Percentiles</h3>

```python
value =np.random.normal(0,0.5,10000) #higher number reacte a normal curve
plt.figure(figsize=[10,4])
plt.hist(value,50)
plt.axvline(np.percentile(value, 25), color='red', label='25%')
plt.axvline(np.percentile(value, 75), color='red', label='75%')
plt.axvline(np.percentile(value, 50), color='yellow', label='middle')
plt.legend(loc='upper right')
plt.show()
```

<body><h3 align='center'>Moments</h3>
<ul>Quantitative measures of the shape of a probability density function
<li>The first moment is the average
<li>The second moment is the variance
<li>The thrid moment is "skew" (γ)
<ul><li>how lopsided a distribution is.
<li> A distribution with a longer tail on the left will be skewed lefet, and have a negative skew.</ul>
<li>The fourth moment is "kurtosis"
<ul><li> How tick is the tail and how sharp is the peak, compared to a normal distribution.
<li> higher peaks have higher kurtosis</ul></ul>
</body>

```python
import scipy.stats as sp
```

```python
#its measure the shape and it doesn't really say where that distribution is exactly
value =np.random.normal(-3,0.5,10000)
plt.figure(figsize=[10,4])
plt.hist(value,50)
plt.axvline(np.mean(value), color='purple', label='first moment')
plt.axvline(np.var(value), color='red', label='second moment')
plt.axvline(sp.skew(value), color='pink', label='third moment')
plt.axvline(sp.kurtosis(value), color='orange', label='fourth moment')

plt.legend(loc='upper right')
plt.show()
```
